```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
whiskey = pd.read_csv("whiskey.csv")
```

```python
whiskey.head()
```

homework-explore about apriori lib
https://rasbt.github.io/mlxtend/user_guide/frequent_patterns/apriori/

```python
#get the most freq itemsets.
#dataset shld be encoded(catg encoding)
freq_itemset = apriori(df = whiskey, min_support=0.1, use_colnames=True)
```

```python
freq_itemset.sample(10)
```

```python
rule = association_rules(freq_itemset, min_threshold=0.5)#by default the metric is 'confidence'
```

```python
rule
```

```python
rule.sort_values(by='lift',ascending=False)
```
