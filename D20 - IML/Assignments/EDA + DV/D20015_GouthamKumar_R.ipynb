## Simple Linear regression co.eff using Algebraic method

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score

def SLE(x,y):
    a=0
    b=0
    n= len(x) 
    a = (np.sum(x**2)*np.sum(y) - (np.sum(x)*(np.sum(x*y)))) / ((n * np.sum(x**2)) - (np.sum(x))**2)
    
    b = ((n*np.sum(x*y) - (np.sum(x)*np.sum(y)))/((n * np.sum(x**2)) - (np.sum(x))**2) )
    
    print("Slope of the line : b: {0},\ny-intercept of the line:  a: {1}".format(b,a))
    return a , b
```

```python
#Predictor and target
x= np.array([1,2,3,4,5])
y= np.array([7,14,15,18,19])
```

```python
#Fitting the model
a, b = SLE(x,y)
```

```python
#plotting
y_line = a + b*x

plt.scatter(x, y, color="red")
plt.plot(x, y_line, color= "green")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Simple Linear Regression");
```

```python
#evalaution of SLE
resiudals = y - y_line
sum_residuals = np.sum(resiudals**2)
#meansq error
n = len(y_line)
mse = sum_residuals/ n 
#root mean sq error
rmse = np.sqrt(mse)
#R2
y_mean = np.mean(y)
SSt = np.sum((y - y_mean)**2) 
R2 = 1- (sum_residuals/SSt)
#Final REsult
print("RMSE:{0}\nR2:{1}".format(rmse,R2))
```

```python
#verifying it with pacakge
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
#we have to change the sahpe
x = x.reshape(-1,1)
#model intialisation and fitting
model = LinearRegression()
model.fit(x,y)
#model predcition
y_pred = model.predict(x)
#Evalaution
mse=mean_squared_error(y,y_pred)
rmse = np.sqrt(mean_squared_error(y, y_pred)) 
r2 = r2_score(y, y_pred)
#result
print("Slope of the line : b: {0},\ny-intercept of the line:  a: {1}".format(model.coef_,model.intercept_))
print("\nRMSE:{0}\nR2:{1}".format(rmse,r2))
```

## Eucledain distance - Vectorised form

```python
x1 = np.array([1,2,3])
x2 = np.array([5,4,3])
```

Eucledian Distance formula:<br/>
d^2 = (x2-x1)^2 

Vectorised form<br/>
d = sqrt((x2-x1)T.(x2-x1))

```python
#subtraction
x2-x1
#xt.x = s^sq
d_sq = np.matmul((x2-x1).T,(x2-x1))

distance = np.sqrt(d_sq)
print("The Eucledian Distance between the vectors x2 and x1 is:",distance)
```

Closed Form Solution:
> b = inverse(HT.H) . (HT.y)

```python
# H matrix = (1,x)
H = np.matrix(x)
y = np.matrix(y)

#H.T . H
HTH = np.matmul((H.T),H)

#Inverse of HT.H
inv_HTH = np.linalg.inv(HTH)

#Let's multiply the baove matrix with HT
final = np.matmul(inv_HTH,H.T)

#Closed form solution
b = np.matmul(final, y)

#solution
print("The intercept a:{0}, \nco.eff b1 for Displacement:{1}, \nco.eff b1 for Horsepower:{2}, \nco.eff b1 for Weight:{3}".format(b[0],b[1],b[2],b[3]))
```

## Graident descent

```python
#for graident descent we have to scale the features to a standard scale
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
x = df[["Displacement","Horsepower","Weight"]]
scaled = scalar.fit_transform(x.values)
scaled
```

```python
scaled_df = pd.DataFrame(scaled, index=x.index, columns=x.columns)
x = scaled_df
x
```

```python
y = df[["MPG"]]
H = sm.add_constant(x)
#create a b vector 
b_ini = np.zeros(shape=[H.shape[1],1])
#for RSS we ahve to check how far it is from mean
b_ini = H.mean()
#first prediction -> H mean
b_ini
```

```python
#Let's convert it to matrix
y = np.matrix(y)
H = np.matrix(H)
b_ini = np.matrix(b_ini)
b_ini = b_ini.T
b_ini
```

```python
#RSS function 
def RSS(H,y,b_ini):
    res = y-np.matmul(H,b_ini)
    rss = -2*(np.matmul(np.transpose(H),res))
    return rss
```

```python
RSS(H,y,b_ini)
```

>Objective: we have to minimise the grad RSS

```python
#Graident descent algorithm

#Intialise the vector
b0 = b_ini
#first grad 
grad = RSS(H,y,b0)
#convergence criteria
conveg = 0.001
alpha = 0.01
i = 1
#we have to make the |RSS| = 0
mag_rss = np.linalg.norm(grad)

while abs(mag_rss) > 1:
    print("The value of Graident:",mag_rss)
    b0 = b0 - ((alpha/np.sqrt(i))*grad)
    i += 1
    grad = rss(H,y,b0)
    mag_rss = np.linalg.norm(grad)
print("The minimum value of the Graident is:{0}\nThe parameters are{1}:".format(mag_rss,b_ini))
```
