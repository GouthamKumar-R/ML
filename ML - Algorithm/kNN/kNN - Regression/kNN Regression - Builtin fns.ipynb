```python
#importing libraries 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('cars.csv')
data.shape
```

```python
#selecting features according to the problem statement
df = data[["MPG","Cylinders","Weight","Horsepower"]]
df.shape
```

```python
df = df[(df[['MPG','Horsepower']] != 0).all(axis = 1)]
df.shape
```

```python
#seperating independent and dependent variables
x = df.drop(['MPG'], axis=1)
y = df['MPG']
x.shape, y.shape
```

```python
#scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
#converting it to df
x = pd.DataFrame(x_scaled, columns = x.columns)
```

```python
# Importing Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y, random_state = 56)
```

```python
#importing KNN regressor and metric mse

from sklearn.neighbors import KNeighborsRegressor as KNN
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import accuracy_score
```

```python
# Creating instance of KNN
reg = KNN(n_neighbors = 13)

# Fitting the model
reg.fit(X_train, y_train)

# Predicting over the Train Set and calculating MSE
test_predict = reg.predict(X_test)
k = mse(test_predict, y_test)
print('Test MSE    ', k )
# print(accuracy_score(test_y,test_predict)*100)
# test_predict
```

```python
def Elbow(K):
  #initiating empty list
    test_mse = []
  #training model for evey value of K
    for i in K:
        #Instance of KNN
        reg = KNN(n_neighbors = i)
        reg.fit(X_train, y_train)
        #Appending mse value to empty list claculated using the predictions
        tmp = reg.predict(X_test)
        tmp = mse(tmp,y_test)
        test_mse.append(tmp)
    
    return test_mse
```

```python
#Defining K range
k = range(1,200) #tried (1,200)

# calling above defined function
test = Elbow(k)
```

```python
# plotting the Curves
plt.plot(k, test)
plt.xlabel('K Neighbors')
plt.ylabel('Test Mean Squared Error')
plt.title('Elbow Curve for test')
```
